```python
def showgray(img):
    fig, ax = plt.subplots()
    ax.imshow(img, cmap='gray')
    ax.spines['top'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.set_xticks([])
    ax.set_yticks([])
    plt.show()
    
def plot4x4(labels, imgs, size=(10, 10)):
    if size:
        plt.figure(figsize=size)
    for i in range(len(imgs)):
        ax = plt.subplot(2,2,i+1)
    #     plt.grid(False)
        plt.plot(imgs[i])
        plt.xlabel(labels[i])
    plt.show()
def plotmxn(m, n, data, is_image, labels=None,  size=(10, 10)):
    if size:
        plt.figure(figsize=size)
    for i in range(len(data)):
        ax = plt.subplot(m,n,i+1)
    #     plt.grid(False)
        if is_image:
            plt.imshow(data[i])
        else:
            plt.plot(data[i][0], data[i][1])
        if labels:
            plt.xlabel(labels[i])
    plt.show()
```

```python
# Ham convolution 2d, khong padding
def conv2d(image, filter):
    # https://stackoverflow.com/questions/43086557/convolve2d-just-by-using-numpy
    sub_shape = filter.shape
    view_shape = tuple(np.subtract(image.shape, sub_shape) + 1) + sub_shape
    strides = image.strides + image.strides
    sub_matrices = np.lib.stride_tricks.as_strided(image, view_shape, strides)
    # Sửa ij thành kl
    # np.einsum('ij,ijkl->ij', filter, sub_matrices)
    return np.einsum('kl,ijkl->ij', filter, sub_matrices)
```

```python
def gauss2D(shape=(3,3), r=1,sigma=0.5, mu = 0):
    """
    2D gaussian mask - should give the same result as MATLAB's
    fspecial('gaussian',[shape],[sigma])
    """
#     m,n = [(ss-1.)/2. for ss in shape]
#     y,x = np.ogrid[-m:m+1,-n:n+1]
#     h = np.exp( -(x*x + y*y) / (2.*sigma*sigma) )
# #     h[ h < np.finfo(h.dtype).eps*h.max() ] = 0
# #     sumh = h.sum()
# #     if sumh != 0:
# #         h /= sumh
    x, y = np.meshgrid(np.linspace(-r,r,shape[0]), np.linspace(-r,r,shape[0]))
    d = np.sqrt(x*x+y*y)
    g = np.exp(-((d-mu)**2 / ( 2.0 * sigma**2 )))/(sigma*math.sqrt(2*math.pi))
    return g

def gauss1D(shape=(3,),sigma=0.5, mu = 0, r=1):
    """
    2D gaussian mask - should give the same result as MATLAB's
    fspecial('gaussian',[shape],[sigma])
    """
#     m,n = [(ss-1.)/2. for ss in shape]
#     y,x = np.ogrid[-m:m+1,-n:n+1]
#     h = np.exp( -(x*x + y*y) / (2.*sigma*sigma) )
# #     h[ h < np.finfo(h.dtype).eps*h.max() ] = 0
# #     sumh = h.sum()
# #     if sumh != 0:
# #         h /= sumh
    x= np.meshgrid(np.linspace(-r,r,shape[0]))[0]
    g = np.exp(-(x - mu)**2 / ( 2.0 * sigma**2 ))/(sigma*math.sqrt(2*math.pi))
    return g
```

```python
def matlab_style_gauss2D(shape=(3,3),sigma=0.5):
    """
    2D gaussian mask - should give the same result as MATLAB's
    fspecial('gaussian',[shape],[sigma])
    """
    m,n = [(ss-1.)/2. for ss in shape]
    y,x = np.ogrid[-m:m+1,-n:n+1]
    h = np.exp( -(x*x + y*y) / (2.*sigma*sigma) )
    h[ h < np.finfo(h.dtype).eps*h.max() ] = 0
    sumh = h.sum()
    if sumh != 0:
        h /= sumh
    return h
```

```python
def linearblur(image, kshape):
    # https://stackoverflow.com/questions/43086557/convolve2d-just-by-using-numpy
    sub_shape = kshape
    view_shape = tuple(np.subtract(image.shape, sub_shape) + 1) + sub_shape
    strides = image.strides + image.strides
    sub_matrices = np.lib.stride_tricks.as_strided(image, view_shape, strides)
    gauss_k = np.empty(sub_matrices.shape)
    for i in range(gauss_k.shape[0]):
        for j in range(gauss_k.shape[1]):
            gauss_k[i,j, :, :] = matlab_style_gauss2D(gauss_k.shape[2:], sigma=0.1 + 5*j/image.shape[1])
#     showgray(gauss_k[0, 0, :, :])
    # Sửa ij thành kl
    # np.einsum('ij,ijkl->ij', filter, sub_matrices)
    return np.einsum('ijkl,ijkl->ij', gauss_k, sub_matrices)
```

```python
# Ham convolution 2d, khong padding
def local_max(image, kshape):
    # https://stackoverflow.com/questions/43086557/convolve2d-just-by-using-numpy
    sub_shape = kshape
    view_shape = tuple(np.subtract(image.shape, sub_shape) + 1) + sub_shape
    strides = image.strides + image.strides
    sub_matrices = np.lib.stride_tricks.as_strided(image, view_shape, strides)
    ln = np.max(np.max(sub_matrices, axis=-1), -1)
    return ln
```

```python
def get_circ_index(n, i):
    bor_len = n - 2*i - 1
    c0 = i * n + i
    c1 = c0 + bor_len
    c2 = bor_len*n + c1
    c3 = c2 - bor_len
    return np.r_[c0:c1, c1:c2:n, c2:c3:-1, c3:c0:-n]

# Ham roll filter
def my_roll(filter, k):
    assert filter.shape[0] == filter.shape[1]
    l = np.ravel(filter.copy())
    n = filter.shape[0]
    for i in range(int(n/2)):
        circ_id = get_circ_index(n, i)
        l[circ_id] = np.roll(l[circ_id], k)
    return l.reshape(n, n)

# Tra ve list tat ca cac xoay vong cua filter
def get_full_roll(filter):
    assert filter.shape[0] == filter.shape[1]
    num_roll = filter.shape[0]**2 - 1
    # rolled_filters = np.empty((*filter.shape, num_roll))
    rolled_filters = []
    for i in range(num_roll):
      # rolled_filters[:, :, i] = my_roll(filter, i)
      rolled_filters.append(my_roll(filter, i))
    return rolled_filters
    
# Ham convolution 2d, khong padding
def conv2d(image, filter):
    # https://stackoverflow.com/questions/43086557/convolve2d-just-by-using-numpy
    sub_shape = filter.shape
    view_shape = tuple(np.subtract(image.shape, sub_shape) + 1) + sub_shape
    strides = image.strides + image.strides
    sub_matrices = np.lib.stride_tricks.as_strided(image, view_shape, strides)
    # Sửa ij thành kl
    # np.einsum('ij,ijkl->ij', filter, sub_matrices)
    return np.einsum('kl,ijkl->ij', filter, sub_matrices)

# conv2d voi nhieu filter va lay max output cua cac filter tren moi pixel
# cac filter phai cung kich thuoc
def compass_conv(image, filters):
    num_filter = len(filters)
    output_shape = tuple(np.subtract(image.shape, filters[0].shape) + 1)
    output = np.empty((*output_shape, num_filter))
    # Convolution
    for i in range(num_filter):
        output[:, :, i] = conv2d(image, filters[i])
    output = np.abs(output)
    # Trả về giá trị lớn nhất trong các rolled filter 
    return np.max(output, axis=2)

kirsh_filter = np.array([
    [5, 5, 5],
    [-3, 0, -3],
    [-3, -3, -3]
])
```

```python
def estimate_blur(image: np.array, threshold: int = 100):
    if image.ndim == 3:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    blur_map = cv2.Laplacian(image, cv2.CV_64F)
    score = np.var(blur_map)
    return blur_map, score, bool(score < threshold)
def pretty_blur_map(blur_map: np.array, sigma: int = 5, min_abs: float = 0.5):
    abs_image = np.abs(blur_map).astype(np.float32)
    abs_image[abs_image < min_abs] = min_abs

    abs_image = np.log(abs_image)
    cv2.blur(abs_image, (sigma, sigma))
    return cv2.medianBlur(abs_image, sigma)
```

```python
bm,_, _ = estimate_blur(image)
showgray(pretty_blur_map(bm, min_abs=0.4))
```
