```python
def showgray(img):
    fig, ax = plt.subplots()
    ax.imshow(img, cmap='gray')
    ax.spines['top'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.set_xticks([])
    ax.set_yticks([])
    plt.show()
    
def plot4x4(labels, imgs):
    plt.figure(figsize=(10, 10))
    for i in range(len(imgs)):
        ax = plt.subplot(2,2,i+1)
    #     plt.grid(False)
        plt.plot(line, imgs[i])
        plt.xlabel(labels[i])
    plt.show()
```

```python
# Ham convolution 2d, khong padding
def conv2d(image, filter):
    # https://stackoverflow.com/questions/43086557/convolve2d-just-by-using-numpy
    sub_shape = filter.shape
    view_shape = tuple(np.subtract(image.shape, sub_shape) + 1) + sub_shape
    strides = image.strides + image.strides
    sub_matrices = np.lib.stride_tricks.as_strided(image, view_shape, strides)
    # Sửa ij thành kl
    # np.einsum('ij,ijkl->ij', filter, sub_matrices)
    return np.einsum('kl,ijkl->ij', filter, sub_matrices)
```

```python
def gauss2D(shape=(3,3),sigma=0.5, mu = 0):
    """
    2D gaussian mask - should give the same result as MATLAB's
    fspecial('gaussian',[shape],[sigma])
    """
#     m,n = [(ss-1.)/2. for ss in shape]
#     y,x = np.ogrid[-m:m+1,-n:n+1]
#     h = np.exp( -(x*x + y*y) / (2.*sigma*sigma) )
# #     h[ h < np.finfo(h.dtype).eps*h.max() ] = 0
# #     sumh = h.sum()
# #     if sumh != 0:
# #         h /= sumh
    x, y = np.meshgrid(np.linspace(-1,1,shape[0]), np.linspace(-1,1,shape[0]))
    d = np.sqrt(x*x+y*y)
    g = np.exp(-((d-mu)**2 / ( 2.0 * sigma**2 )))/(sigma*math.sqrt(2*math.pi))
    return g

def gauss1D(shape=(3,),sigma=0.5, mu = 0, r=1):
    """
    2D gaussian mask - should give the same result as MATLAB's
    fspecial('gaussian',[shape],[sigma])
    """
#     m,n = [(ss-1.)/2. for ss in shape]
#     y,x = np.ogrid[-m:m+1,-n:n+1]
#     h = np.exp( -(x*x + y*y) / (2.*sigma*sigma) )
# #     h[ h < np.finfo(h.dtype).eps*h.max() ] = 0
# #     sumh = h.sum()
# #     if sumh != 0:
# #         h /= sumh
    x= np.meshgrid(np.linspace(-r,r,shape[0]))[0]
    g = np.exp(-(x - mu)**2 / ( 2.0 * sigma**2 ))/(sigma*math.sqrt(2*math.pi))
    return g
```

```python
def matlab_style_gauss2D(shape=(3,3),sigma=0.5):
    """
    2D gaussian mask - should give the same result as MATLAB's
    fspecial('gaussian',[shape],[sigma])
    """
    m,n = [(ss-1.)/2. for ss in shape]
    y,x = np.ogrid[-m:m+1,-n:n+1]
    h = np.exp( -(x*x + y*y) / (2.*sigma*sigma) )
    h[ h < np.finfo(h.dtype).eps*h.max() ] = 0
    sumh = h.sum()
    if sumh != 0:
        h /= sumh
    return h
```

```python
def linearblur(image, kshape):
    # https://stackoverflow.com/questions/43086557/convolve2d-just-by-using-numpy
    sub_shape = kshape
    view_shape = tuple(np.subtract(image.shape, sub_shape) + 1) + sub_shape
    strides = image.strides + image.strides
    sub_matrices = np.lib.stride_tricks.as_strided(image, view_shape, strides)
    gauss_k = np.empty(sub_matrices.shape)
    for i in range(gauss_k.shape[0]):
        for j in range(gauss_k.shape[1]):
            gauss_k[i,j, :, :] = matlab_style_gauss2D(gauss_k.shape[2:], sigma=0.1 + 5*j/image.shape[1])
#     showgray(gauss_k[0, 0, :, :])
    # Sửa ij thành kl
    # np.einsum('ij,ijkl->ij', filter, sub_matrices)
    return np.einsum('ijkl,ijkl->ij', gauss_k, sub_matrices)
```
